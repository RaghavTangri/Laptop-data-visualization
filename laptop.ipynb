```python
#importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('laptopData.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
#shape of data
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# dropping unnamed column
df.drop(columns='Unnamed: 0',inplace = True)
```

```python
#getting sum of duplicates
df.duplicated().sum()
```

```python
#dropping duplicates then resetting index
df.drop_duplicates(inplace=True,ignore_index=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
#fill missing price values
mean_price = df['Price'].mean()
df['Price'].fillna(mean_price, inplace=True)
df.isnull().sum().sum()
```

```python
df['ScreenResolution'].mode()
```

```python
# filling missing screenresolution value
print(df['ScreenResolution'].isnull().sum())
mode_screen = df['ScreenResolution'].mode()[0]
df['ScreenResolution'].fillna(mode_screen,inplace=True)
print(df['ScreenResolution'].isnull().sum())
```

```python
df.isnull().sum().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
# getting count of comapanies value
df['Company'].value_counts()
```

```python
#unique values in weight
df.Weight.unique()
```

```python
df.ScreenResolution.unique()
```

```python
# splitting screen resolution
df['ScreenResolution'] = df['ScreenResolution'].str.split().str[-1]
df.ScreenResolution.unique()
```

```python
df.Cpu.unique()
```

```python
#cpu

df.Cpu.unique()
```

```python
df.Memory.unique()
```

```python
#removing rows where ?
df.drop(index=df[df['Memory']=='?'].index[0],inplace=True)
```

```python
df.Memory.unique()
```

```python
df.Inches.unique()
```

```python
# removing ? from inches
df.drop(index=df[df['Inches']=='?'].index[0],inplace=True)
```

```python
df.Inches.unique()
```

```python
df['Inches']=df['Inches'].astype('float')
```

```python
df.info()
```

```python
sns.boxplot(data=df, x='Price', y=None, hue=None)
```

```python
#highest price laptop
plt.bar(df['Company'], df['Price'])
plt.xlabel("Company")
plt.xticks (rotation='vertical')
plt.ylabel("Price")
plt.title("highest price laptop")
plt.show()
```

```python
plt.bar(df['Company'], df['ScreenResolution'])
plt.xticks (rotation='vertical')
plt.title("Highest Screen resolution with company")
plt.show()
```

```python
sns.scatterplot(data=df,x="Price",y="ScreenResolution",hue="Price")
plt.title("Scatter Plot showing price based on screen resolution")
plt.show()
```

```python
df["Inches"].plot(kind = 'hist')
plt.xlabel("Inches")
```

```python
sns.histplot(data=df,x='Price',y='Company')
plt.title("Company vs price")
```
